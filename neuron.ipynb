```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid activation
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Derivative of sigmoid (used in backprop)
def sigmoid_derivative(z):
    return sigmoid(z) * (1 - sigmoid(z))

# Data
x = np.array([0, 1])
y = np.array([0, 1])

# Parameters
w = np.random.randn()
b = 0.0
learning_rate = 0.1
epochs = 1000

losses = []

# Training loop
for epoch in range(epochs):
    # Forward pass
    z = w * x + b
    y_pred = sigmoid(z)

    # Loss (mean squared error)
    loss = np.mean((y - y_pred) ** 2)
    losses.append(loss)

    # Gradients
    dz = 2 * (y_pred - y) * sigmoid_derivative(z)
    dw = np.sum(dz * x) / len(x)
    db = np.sum(dz) / len(x)

    # Update
    w -= learning_rate * dw
    b -= learning_rate * db

    if epoch % 100 == 0:
        print(f"Epoch {epoch} | Loss: {loss:.4f} | w: {w:.4f} | b: {b:.4f}")

print(f"\nFinal neuron: y = sigmoid({w:.2f} * x + {b:.2f})")

# Plot predictions
x_test = np.linspace(-1, 2, 100)
y_test = sigmoid(w * x_test + b)

plt.plot(x_test, y_test, label="Neuron Output")
plt.scatter(x, y, color="red", label="Training Data")
plt.title("Single Neuron Prediction")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
