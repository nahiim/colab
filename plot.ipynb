```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Training data
x = np.array([[1], [2], [3], [4]])
y = np.array([2, 4, 6, 8])

# Step 2: Create and train the model
model = LinearRegression()
model.fit(x, y)

# Step 3: Predict a new value
predicted = model.predict([[5]])
print("Prediction for 5:", predicted[0])

# Step 4: Plot data + model
x_test = np.linspace(0, 6, 100).reshape(-1, 1)  # 100 points from 0 to 6
y_pred_line = model.predict(x_test)

plt.scatter(x, y, color="blue", label="Training Data")
plt.plot(x_test, y_pred_line, color="red", label="Model Prediction")
plt.scatter([5], predicted, color="green", label="Prediction at x=5", zorder=5)
plt.legend()
plt.title("Linear Regression Model")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
